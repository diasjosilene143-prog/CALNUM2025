#Atividade ‚Äì Implementa√ß√£o do M√©todo de Jacobi em Python

Enunciado
Voc√™ dever√° implementar, em Python, o m√©todo de Jacobi para resolver sistemas lineares do tipo ùê¥ùë•=ùëè.

Parte 1 ‚Äì Fun√ß√£o do m√©todo
Crie uma fun√ß√£o em Python que:

receba a matriz ùê¥ , o vetor ùëè , um chute inicial, a toler√¢ncia e o n√∫mero m√°ximo de itera√ß√µes;

calcule as itera√ß√µes do m√©todo de Jacobi;

retorne:

a solu√ß√£o aproximada,o n√∫mero de itera√ß√µes realizadas,e mostre na tela, passo a passo, cada itera√ß√£o (valores de ùë• e erro).

Parte 2 ‚Äì Teste do algoritmo
Na parte principal do programa (fora da fun√ß√£o):

Defina pelo menos dois sistemas lineares diferentes (matrizes ùê¥ e vetores ùëè );

Para cada sistema, chame a fun√ß√£o criada e exiba o resultado final.

Requisitos
O c√≥digo deve estar organizado com a fun√ß√£o em um bloco e as chamadas da fun√ß√£o em outro.

Em cada itera√ß√£o, mostrar algo como:

Itera√ß√£o 1: x = [ ... ] erro = ...

Itera√ß√£o 2: x = [ ... ] erro = ...

ou uma tabela com os resultados de cada intera√ß√£o

Objetivo da atividade
Compreender o funcionamento do m√©todo iterativo de Jacobi, implementando o algoritmo e analisando sua converg√™ncia atrav√©s das aproxima√ß√µes sucessivas.

```python
import numpy as np #Importa a biblioteca que facilita opera√ß√µes com matrizes

# BLOCO 1 ‚Äî Fun√ß√£o do Jacobi

def jacobi(A, b, x0, tol, max_iter):
    #Utilizando o float nas vari√°veis evita problemas com aproxima√ß√µes.
    A = A.astype(float)
    b = b.astype(float)
    x = x0.astype(float)
    n = len(b) #Quantidade de vari√°veis do meu sistema

    # Check for diagonal dominance and potential division by zero for robustness
    is_strictly_diagonally_dominant = True
    for i in range(n):
        if A[i, i] == 0:
            raise ValueError(f"Diagonal element A[{i},{i}] is zero, which would lead to division by zero. Cannot proceed with Jacobi method.")
        sum_off_diag = sum(abs(A[i, j]) for j in range(n) if j != i)
        if abs(A[i, i]) < sum_off_diag:
            is_strictly_diagonally_dominant = False

    if not is_strictly_diagonally_dominant:
        print("Warning: The matrix is not strictly diagonally dominant. Convergence of the Jacobi method is not guaranteed.")

    tabela = []  # Lista para armazenar os dados que ser√£o utilizados na tabela depois

    # adiciona primeiro estado (k=0)
    erro0 = np.max(np.abs(x))  # erro inicial s√≥ pra mostrar na tabela
    tabela.append((0, x.copy(), erro0)) #Salva a primeira linha da tabela

#Utiliza-se o "for" para que o programa aconte√ßa em loop at√© o "crit√©rio de parada"
    for k in range(1, max_iter + 1):
        x_new = np.zeros_like(x) #Cria um vetor novo que armazena os valores para a intera√ß√£o seguinte

#Da mesma forma que usamos o for antes para loop, utilizamos aqui para que ele percorra cada equa√ß√£o
        for i in range(n):
            soma = sum(A[i, j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - soma) / A[i, i]

        erro = np.max(np.abs(x_new - x)) #Calcula o erro como o maior valor da diferen√ßa.

        tabela.append((k, x_new.copy(), erro)) #Salva a intera√ß√£o realizada na tabela criada.

        if erro < tol:
            return x_new, k, tabela

#Para que ocorra uma nova intera√ß√£o com o vetor novo, uitlizamos:
        x = x_new

    return x, max_iter, tabela #Ao atingir o m√°ximo de intera√ß√µes, retorna assim mesmo.
```

```python
import numpy as np

#BLOCO 2 - Impress√£o da Tabela organizada
#T√≠tulos para sa√≠da do programa
def imprimir_saida(A, b, solucao, tabela):
    print("M√©todo de Jacobi")
    print("Matriz Ampliada")

    # imprime cada linha da matriz ampliada A junto com b ao lado
    for i in range(len(b)):
        linha = " ".join(f"{A[i,j]:7.4f}" for j in range(len(b)))
        print(f"{linha} {b[i]:7.4f}") #Imprime A e b juntos

    print("\n k       x1       x2       x3       Erro") #Dados da tabela

    for k, x, erro in tabela:
        print(f"{k:2d}  {x[0]:7.4f}  {x[1]:7.4f}  {x[2]:7.4f}  {erro:7.4f}") #Imprime todas as intera√ß√µes que ocorreram no loop, na tabela.

    print("\nSolu√ß√£o do sistema")
    for i, xi in enumerate(solucao, start=1):
        print(f"   {xi:.4f}", end="")
    print("\n")
```
